이상치 정제하기

```r
outlier <- data.frame(sex = c(1,2,1,3,2,1),
                     score = c(5,4,3,4,2,6))
outlier
```

```r
table(outlier$sex)
```

```r
table(outlier$score)
```

```r
outlier$sex <- ifelse(outlier$sex == 3, NA, outlier$sex)
outlier
```

```r
outlier$score <- ifelse(outlier$score >5, NA, outlier$score)
outlier
```

```r
library(dplyr)
outlier %>%
    filter(!is.na(sex) & !is.na(score)) %>%
    group_by(sex) %>%
    summarize(mean_score = mean(score))
```
